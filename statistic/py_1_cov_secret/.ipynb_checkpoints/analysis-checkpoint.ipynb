### import module

```python
import numpy as np
import pandas as pd
```

### get data

```python
path = "data.csv"

data = pd.read_csv(path)
data["date"] = pd.to_datetime(data["date"])
data.set_index("date", inplace=True)
```

```python
data.describe()
```

## 時系列データ
1. ARモデル（自己回帰モデル）  
    ある時点のデータを過去のデータを用いて回帰するモデル
1. MAモデル（移動平均モデル）

## x axis: 2020/10/1 ～ 2020/10/31 space(5)

```python
st = data.index[0]
ed = data.index[-1]
x_axis = pd.to_datetime(
    np.linspace(st.value, ed.value, 5))
```

```python
import matplotlib.pyplot as plt

plt.plot(
    data.index, data["Sample"], color='r', label="lineer", alpha=0.6
)
plt.plot(
    data.index, data["Sample"].interpolate(method="cubic", order=2), color="y", label="spline", alpha=0.8
)
plt.xticks(x_axis)
plt.show()
```

```python
x_axis
```

```python
plt.bar(data.index, data["Sample"])
plt.xticks(x_axis)
plt.show()
```
