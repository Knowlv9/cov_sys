##### $\mathscr{Q.1}$
&emsp;$ x_1, ..., x_N, y_1, ..., y_N \ni \mathbb{R}$について、$\mathit{L}: \displaystyle \sum_{i=1}^N (y_i - \beta_0 - \beta_1x_i)^2$ を最小にする$\beta_0, \beta_1 \ni \mathbb{R}$を
$\;\hat{\beta_0}, \:\hat{\beta_0}\;$とおくとき、
以下の等式を示せ. <br>
ただし、$ \bar{x} $ および $ \bar{y} $ は
$\displaystyle \frac{1}{N}\displaystyle \sum_{i=1}^N x_i$ および 
$\displaystyle \frac{1}{N}\displaystyle \sum_{i=1}^N y_i$ で定義されるものとする.<br>
<br>
(a) $\hat{\beta_0} + \hat{\beta_1}\bar{x} = \bar{y}$<br>
(b) $x_1, ..., x_N$ がすべて等しくないとき、
$$
\hat{\beta_1} = \frac{\displaystyle \sum_{i=1}^N(x_i - \bar{x})(y_i - \bar{y})}{\displaystyle \sum_{i=1}^N(x_i - \bar{x})^2}
$$

### (a)
$$
\mathit{L} = \displaystyle \sum_{i=1}^N (y_i - \beta_0 - \beta_1x_i)^2
$$
より、
$ \mathit{l} = (y_i - \beta_0 - \beta_1x_i)^2 $ とすると、<br>
$ \beta_0 $についてまとめると、
$ \mathit{l} = \beta_0^2 - 2(x_i - \beta_1x_i)\beta_0 + (y_i - \beta_1x_i)^2 $ から、<br>
$ \beta_0 $を偏微分して、0とおくと、
$$
\begin{align}
\displaystyle \frac{ \partial l}{\partial \beta_0} = 2\beta_0 - 2(y_i - \beta_1x_i) &= 0 \\
2\beta_0 &= 2(y_i - \beta_1x_i)\\
\beta_0 &= y_i - \beta_1x_i
\end{align}
$$

$\hat{\beta_0} + \hat{\beta_1}\bar{x} = \bar{y}$において、
$ \hat{\beta_0} $は切片、$ \hat{\beta_1} $は傾きであるから、  $\mathit{L}$ を$\:\beta_0\:$で偏微分して0とおくと、
$$
\begin{align}
\frac{ \partial L }{\partial \beta_0 } = -2 \displaystyle \sum_{i=1}^N (y_i - \beta_0 - \beta_1x_i) &= 0\\
\hat{\beta_0} &= \bar{y} - \hat{\beta_1}\bar{x}\\
\bar{y} &= \hat{\beta_0} + \hat{\beta_1}\bar{x}
\end{align}
$$
