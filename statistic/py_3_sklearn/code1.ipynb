```python
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score
import pandas as pd
```

## データセットの準備
    data.csv:
        1列目：　目的変数
        2列目以降：説明変数

```python
data = np.loadtxt("data.csv", delimiter=',', dtype=float)
labels = data[:, 0:1]
featurs = preprocessing.minmax_scale(data[:, 1:])
x_train, x_test, y_train, y_test = train_test_split(featurs, labels.ravel(), test_size=0.3)
```

```python
print(x_train)
```

### select model

```python
clf = svm.SVC(kernel='rbf', C=10, gamma=0.1)
```

### learning

```python
clf.fit(x_train, y_train)
```

### predict

```python
predict = clf.predict(x_test)
```

### accuracy, precicision, recall
    accuracy: 「真値」にどれだけ近いか
    precision: 精度. 複数回の測定値間でのばらつきの度合い

```python
print(
    "score\n"
    "\taccuracy: %s\n\tprecision: %s\n\trecall: %s" % (
        accuracy_score(y_test, predict),
        precision_score(y_test, predict),
        recall_score(y_test, predict))
)
```
