```python
# -*- coding: utf-8 -*-

import csv
import pandas as pd
import numpy as np
from scipy import stats
```

update csv data in every 5 min.

```python
filename = "./../../task/app_1/cov_system/src/data/download/sample_5000_1.csv"
df = pd.read_csv(filename, index_col=0)

f_sex = lambda x: 0 if x == '男' else 1
df['性別'] = df['性別'].map(f_sex)
df.head()
```

```python
# 材料 0:鼻腔, 1:唾液
matList = np.random.choice([0, 1], size=len(df), p=[0.4, 0.6])
df['材料'] = matList
expireList = np.random.choice([0, 1, 2], size=len(df), p=[0.3, 0.4, 0.3])
df['報告期限'] = expireList
```

```python
hp_dir = "./../../task/app_1/cov_system/src/data/hp_names.csv"
hp_df = pd.read_csv(hp_dir)
hp_df.rename(columns={'Unnamed: 0':'name'}, inplace=True)
hp_df.iloc[0]
```

```python
l = 4
sc = 2.45
X = np.arange(start=1, stop=80, step=1)
norm = stats.norm.pdf(x=X, loc=l, scale=sc)
c = 1 - sum(norm)
hpNumP = list(norm)
hpNumP.insert(0, c)
# print(hpNumP)
hpList = np.random.choice(list(range(1, 81)), size=len(df), p=hpNumP)
print(hpList[:100])
```

```python
hp_names = []
office_list = []

arr = list(range(0, 205))
def reset_j(j, i, f):
    if j == 205: f = False
    j = np.random.choice(arr, size=1, p=[1/len(arr)]*len(arr))[0]
    return j, f

s = 0
j = 0
f = True
for i in hpList:
    if s+i > len(df): break
    s += i
#     if j > 200: print(j, f)
    if j > len(hp_df)-1 or f is False:
        j, f = reset_j(j, i, f) 
    hp = [hp_df.iloc[j]['name']] * i
    office = [hp_df.iloc[j]['district']] * i
    hp_names.extend(hp)
    office_list.extend(office)
    j += 1

s2 = 5000 - len(hp_names)
j = np.random.choice(arr, size=1, p=[1/len(arr)]*len(arr))[0]
hp = [hp_df.iloc[j]['name']] * (s2)
hp_names.extend(hp)
office = [hp_df.iloc[j]['district']] * (s2)
office_list.extend(office)
# print(len(hp_names))
df['病院名'] = hp_names
df["営業所"] = office_list
```
