```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# np.random.seed(0)
# 等差数列を生成
l = 4
s = 2.5
X = np.arange(start=1, stop=80, step=1)
# pdfで確率密度関数を生成
norm_pdf = stats.norm.pdf(x=X, loc=l, scale=s)

x = 60
y = stats.norm.pdf(x=x, loc=l, scale=s)
f = 1 - sum(norm_pdf)
li = list(norm_pdf)
li.insert(0, f)
```

```python
X = np.random.choice(list(range(1, 81)), size=1000, p=li)
# alp_li = [chr(ord('a') + i) for i in range(26)]
alp_p = [1/36]* 21
a = (1-sum(alp_p)) / 5
for i in [0, 4, 8, 14, 20]:
    alp_p.insert(i, a)
# print(sum(alp_p))
# alp_p
```

```python
i = 0
hp_list = []
name_dic = {}
alp_li = [chr(ord('a') + i) for i in range(26)]
hps = ["hospital", "clinic", "Clinic", "Municipal hospital"]
def make_hp(n):
    name_len = np.random.choice(list(range(3, 6)), size=1, p=[1/3]*3)
    name = np.random.choice(alp_li, size=name_len[0], p=alp_p)
    
    hp = np.random.choice(hps, size=1, p=[0.4, 0.45, 0.1, 0.05])
    hp_name = ''.join(name) + ' ' + hp[0]
    return hp_name

d = ['Y', 'W', 'V', 'S', 'L', 'B', 'R', 'K']
p_d = [0.175, 0.15, 0.125, 0.075, 0.125, 0.125, 0.125, 0.1]
def get_district():
    o = np.random.choice(d, size=1, p=p_d)
    return o[0]

for x in X:
    if i+x > 1000: break
    name = make_hp(x)
    hp_list += [name] * x
    name_dic[name] = {
        "district": get_district()
    }
    i += x
hp_list += [make_hp(1000 - len(hp_list))] * (1000 - len(hp_list))
```

```python
name_dic = pd.DataFrame(name_dic).T
name_dic.to_csv('./hp_names_2.csv')
```
