## spline complement

```python
import datetime as dt
import pandas as pd
dates = pd.to_datetime(np.linspace(
    pd.Timestamp(2020, 10, 1).value, pd.Timestamp(2020, 10, 31).value, 12
))
x = [int(dates[i].strftime('%d')) for i in range(0, len(dates))]
y = np.random.randn(12)
```

```python
plt.plot(x, y)
```

```python
cx = pd.to_datetime(np.linspace(
    pd.Timestamp(2020, 10, 1).value, pd.Timestamp(2020, 10, 31).value, len(y)*5
))
cx = [cx[i].strftime('%d') for i in range(0, len(cx))]
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind="cubic")
cy = f(cx)
plt.plot(x, y, color="blue", label="liner", alpha=0.7)
plt.plot(cx, cy, color="red", label="cubic", alpha=0.7)
plt.xticks(np.arange(1, 31, 5))
plt.legend()
plt.show()
```
